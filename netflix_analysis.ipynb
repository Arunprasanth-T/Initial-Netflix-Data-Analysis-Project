What is the distribution of TV Shows by number of seasons?

```python
tv = df[df["type"] == "TV Show"].copy()
season_count = tv['duration'].value_counts().sort_index()
```

```python
plt.figure(figsize=(10,5))

season_count.plot(
    kind="bar",
    color="skyblue"
)

plt.title("Distribution of TV Shows by Number of Seasons")
plt.xlabel("Number of Seasons")
plt.ylabel("Number of TV Shows")

plt.xticks(rotation=45)

plt.show()
```

Most TV Shows on Netflix have only 1 Season, while the number of shows decreases as the number of seasons increases. This indicates that Netflix's catalog contains many limited series or newer shows rather than long-running series.

What is the average number of seasons of TV Shows on Netflix?

```python
tv["duration"] = tv["duration"].str.replace(" Seasons", "", regex=False)
tv["duration"] = tv["duration"].str.replace(" Season", "", regex=False)
tv["duration"] = tv["duration"].astype(int)
```

```python
average_seasons = tv["duration"].mean()

print("Average Number of Seasons:", round(average_seasons,2))
```

```python
plt.figure(figsize=(8,5))

tv["duration"].hist(bins=10, edgecolor="black")

plt.title("Distribution of Number of Seasons")
plt.xlabel("Number of Seasons")
plt.ylabel("Frequency")

plt.show()
```

The average TV Show on Netflix has approximately 1–2 seasons, indicating that Netflix's catalog is dominated by limited series and shorter-running shows rather than long-running television series.

```python
df_director = df.copy()
df_director["director"] = df_director["director"].str.split(",")
df_director = df_director.explode("director")
df_director["director"] = df_director["director"].str.strip()
df_director = df_director[df_director["director"] != "Unknown"]
top_directors = df_director["director"].value_counts().head(10)
top_directors
```

```python

plt.figure(figsize=(10,5))

top_directors.plot(
    kind="bar",
    color="purple"
)

plt.title("Top 10 Directors on Netflix")
plt.xlabel("Director")
plt.ylabel("Number of Titles")

plt.xticks(rotation=45, ha="right")

plt.show()
```

Netflix can continue strengthening partnerships with high-performing directors by acquiring more of their content or collaborating on original productions. Recommending movies based on users' favorite directors can also improve personalization and increase watch time.
